## 通过多普勒频移谱生成用于训练卷积网络的数据

```python
import os
import tensorflow as tf
from PIL import Image
import numpy as np
import matplotlib.pyplot as plt
import pickle


set_path= 'pic\\stft_pic'
classes=[]
labels=[]
filenames=os.listdir(set_path)

mapping={}


for item in os.listdir(set_path):
    temp_class=item.split('_')[0]
    classes.append(temp_class)
    
enum=list(set(classes))

for i in range(len(enum)):
    mapping[i]=enum[i]

print(mapping)
for item in classes:
    labels.append(enum.index(item))
```

```python
images=[]
for image in filenames:
    image=set_path+'\\'+image
    print('\r writing in',image,flush=True)
    img = Image.open(image)
    img = img.resize((224,224))
    img = np.array(img)
    images.append(img)

images=np.array(images,dtype=np.float32)
labels=np.array(labels,dtype=np.int32)

print(images.shape)
print(labels)

data={
    'image':images,
    'label':labels,
    'mapping':mapping
}

with open('data.data','wb') as f:
    pickle.dump(data,f)
```
